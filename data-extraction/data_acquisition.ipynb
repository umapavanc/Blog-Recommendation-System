```python
import requests
import pandas as pd
pd.set_option('display.max_colwidth', 100)
import numpy as np
from bs4 import BeautifulSoup
```

```python
base_url = "https://bair.berkeley.edu"
archive_ext = "/blog/archive/"

data = requests.get(base_url + archive_ext)
if int(data.status_code) == 200:
    print(data.headers['Content-Type'])
    print(data.text) # data.content both these tags does the same job
```

```python
soup = BeautifulSoup(data.text)
soup
```

```python
berkeley_dict = {}
i = 0
for link in soup.find_all('li'):
    #print(link.find('a').get('href'))
    #print(link.find('a').text)
    berkeley_dict[i] = {'title' : link.find('a').text, 'URL' : link.find('a').get('href')}
    i = i+1
berkeley_df = pd.DataFrame.from_dict(berkeley_dict, orient = 'index')
print(berkeley_df.info(), "\n")
berkeley_df.head()
```

```python
def av_requests(category, url=None):
    if category:
      url = "https://www.analyticsvidhya.com/blog/category/" + category
    print('url: ', url)
    response = requests.get(url)
    if int(response.status_code) == 200:
        soup = BeautifulSoup(response.text)
    return soup
```

```python
def av_text_scrapper(category, av_dict = None, soup = None):
    if av_dict is None:
        av_dict = {}

    index = len(av_dict)
    if soup is not None:
        for element in soup.find_all('div', class_ = "list-card-content"):
            for child in element.children:
                # Check if the child is a tag
                if child.name == 'a':
                    av_dict[index] = {'title': child.text.strip(), 'URL': child.get('href'), 'category': category}
                    index = index + 1
    return av_dict
```

```python
total_av = {}
for category in {'computer-vision', 'machine-learning'}:
  av_soup = av_requests(category)
  no_pages_links = av_soup.find_all('a', class_ = "page-numbers")
  # print(no_pages_links)
  next_link = no_pages_links[0].get('href')
  total_pages = no_pages_links[1].text
  #no_pages_links, next_link, total_pages

  av_dict = av_text_scrapper(category, soup = av_soup)
  for i in range(2, int(total_pages)+1):
      # print(i)
      soup = av_requests(None, next_link)
      if i == int(total_pages):
          av_dict = av_text_scrapper(category, av_dict, soup)
          break
      else:
          av_dict = av_text_scrapper(category, av_dict, soup)
          no_pages_links = soup.find_all('a', class_ = "page-numbers")
          if i == 2:
              next_link = no_pages_links[2].get('href')
          else:
              next_link = no_pages_links[3].get('href')
  total_av.update({category: av_dict})
```

```python
def ref_requests(category, url=None):
    if category:
      url = "https://www.refinery29.com/en-us/" + category
    else:
      exit
    print(url)
    response = requests.get(url)
    if int(response.status_code) == 200:
        soup = BeautifulSoup(response.text)
    return soup
```

```python
ref_dict = {}
ref_base_url = "https://www.refinery29.com"
for category in ['work-and-money', 'beauty', 'money-diary', 'horoscope', 'fashion', 'entertainment', 'gg-good-game', 'movies', 'cooking-tips']:
  ref_soup = ref_requests(category)
  index = len(ref_dict)
  if ref_soup:
    cards = ref_soup.find_all('div', class_ = "card")
    for card in cards:
      ref_dict[index] = {'title': card.find('div', {'class': 'title'}).text, 'URL': ref_base_url + card.a.get('href'), 'category': category}
      index = index + 1
```

```python
ref29_df = pd.DataFrame.from_dict(ref_dict, orient='index')
ref29_df.to_csv("ref29_data.csv")
```
